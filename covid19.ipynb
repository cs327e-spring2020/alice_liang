```python
%%bigquery
create table covid_modeled.cases as
select distinct date, daily_confirmed_cases, confirmed_cases, daily_deaths, deaths, geo_id, country_territory_code
from `ec-covid-19.covid_staging.covid_19_geographic_distribution_worldwide`
```

```python
%%bigquery
create table covid_modeled.geography as
select distinct countries_and_territories, geo_id, country_territory_code, pop_data_2018
from `ec-covid-19.covid_staging.covid_19_geographic_distribution_worldwide`
```

```python
%%bigquery
create table covid_modeled.mobility as
select country_region_code, country_region, date, retail_and_recreation_percent_change_from_baseline as retail_and_recreation, grocery_and_pharmacy_percent_change_from_baseline as grocery_and_pharmacy,parks_percent_change_from_baseline as parks, transit_stations_percent_change_from_baseline as transit_stations, workplaces_percent_change_from_baseline as workplaces, residential_percent_change_from_baseline as residential
from `ec-covid-19.covid_staging.mobility_report`
```

# Dataset 1 Queries

```python
%%bigquery
select date, confirmed_cases, deaths, countries_and_territories
from covid_modeled.cases c join covid_modeled.geography g
on c.geo_id = g.geo_id
where countries_and_territories = 'China'
order by date desc
limit 20
```

# Cross Dataset Queries

```python
%%bigquery
SELECT distinct c.date, g.countries_and_territories, c.confirmed_cases, c.deaths
FROM `ec-covid-19.covid_modeled.cases` as c 
JOIN `ec-covid-19.covid_modeled.geography` as g on c.geo_id = g.geo_id
WHERE g.countries_and_territories IN (SELECT country_region FROM `ec-covid-19.covid_modeled.mobility` WHERE retail_and_recreation < 0)
ORDER BY c.date desc
limit 20
```
